```python
import pandas as pd
import numpy as np
import time
from google.colab import drive
```

```python
def to_date_time(col):
    return pd.to_datetime(col,errors='coerce', format='%m/%d/%Y %H:%M')
```

```python
drive.mount('/content/drive')
input_dataframe = pd.read_csv("/content/drive/MyDrive/DM-Dataset/PRODUCTINSTANCEREDUCED.csv")

input_dataframe['DOCUMENTDATE'] = to_date_time(input_dataframe['DOCUMENTDATE'])
input_dataframe['REACHMENTDATE'] = to_date_time(input_dataframe['REACHMENTDATE'])
```

```python
def create_table(processing_dataframe, processing_column, processing_column_name):    
    res = []

    for record in processing_dataframe:
        processing_rows = input_dataframe[processing_column == record]
        products_ids = processing_rows.M_TAG_ID.unique()
        products_ids_counts = processing_rows.M_TAG_ID.value_counts()

        for product_id in products_ids:
            row_dataframe = {}

            transfer_count = products_ids_counts[product_id]
            processing_row = processing_rows[processing_rows['M_TAG_ID'] == product_id]

            document_date = processing_row.DOCUMENTDATE.max()
            reachment_date = processing_row.REACHMENTDATE.min()
            transfer_period = document_date - reachment_date

            product_value = processing_row.PRIMALVALUE.mean()

            row_dataframe[processing_column_name] = [record]
            row_dataframe["Product ID"] = [product_id]
            row_dataframe["Product Transfer Count"] = [transfer_count]
            row_dataframe["Product Transfer Period"] = [transfer_period]
            row_dataframe["Product Value"] = [product_value]
            res.append(row_dataframe)

    return res
```

```python
holdings = input_dataframe.ACCT_AC_HOLDING_ID.unique()
assistance = input_dataframe.AD_ORG_REF_ID.unique()
beneficiaries = input_dataframe.COM_BPARTNER_ID.unique()
```

```python
holdings_df = create_table(holdings, input_dataframe.ACCT_AC_HOLDING_ID, "Holding")
result_dataframe = pd.DataFrame.from_dict(holdings_df)
print(result_dataframe)
```

```python
assistance_df = create_table(assistance, input_dataframe.AD_ORG_REF_ID, "Assistance")
result_dataframe = pd.DataFrame.from_dict(assistance_df)
print(result_dataframe)
```

```python
beneficiaries_df = create_table(beneficiaries, input_dataframe.COM_BPARTNER_ID, "Beneficiary")
result_dataframe = pd.DataFrame.from_dict(beneficiaries_df)
print(result_dataframe)
```
